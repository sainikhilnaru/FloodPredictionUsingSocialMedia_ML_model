```python
crs = {'init':'epsg:4326'}
geo_df = geopandas.GeoDataFrame(df, crs=crs, geometry = geometry)
geo_df.head()
```

```python
fig,ax = plt.subplots(figsize = (50,50))
streetMap.plot(ax = ax, color = 'grey')
```

```python
fig,ax = plt.subplots(figsize = (30,30))

streetMap.plot(ax = ax , color ='grey' )
#geo_df[geo_df['total reports'] == 0 ].plot(ax = ax, markersize = 5, color = 'black', marker = 'o', label= 'Flood reports = 0' )
geo_df[geo_df['total reports'].between(1,5,inclusive = False) ].plot(ax = ax, markersize = 10, color = 'blue', marker = 'D', label= ' 0 > Flood reports > 5' )
geo_df[geo_df['total reports'].between(6,10,inclusive = False) ].plot(ax = ax, markersize = 15, color = 'green', marker = '+', label= ' 5 > Flood reports > 10' )
geo_df[geo_df['total reports'].between(11,15,inclusive = False) ].plot(ax = ax, markersize = 20, color = 'yellow', marker = 'x', label= ' 10 > Flood reports > 15' )
geo_df[geo_df['total reports'].between(16,20,inclusive = False) ].plot(ax = ax, markersize = 25, color = 'orange', marker = '^', label= ' 15 > Flood reports > 20' )
geo_df[geo_df['total reports'] > 20 ].plot(ax = ax, markersize = 30, color = 'red', marker = 's', label= 'Flood reports  > 20' )
plt.legend(prop = {'size' : 30})
#c parameter , s parameter, create a color bar/map , don't bin continuous data
```
